```python
import pandas as pd
from matplotlib import pyplot as plt
pd.set_option('display.width', 1000)
```

```python
instacart_orders = pd.read_csv('instacart_orders.csv', sep=';')
products = pd.read_csv('products.csv', sep=';')
order_products = pd.read_csv('order_products.csv', sep=';')
aisles = pd.read_csv('aisles.csv', sep=';')
departments = pd.read_csv('departments.csv', sep=';')
```

```python
pedidos = order_products.groupby("order_id")
productos_por_pedido = pedidos['product_id'].size()
distribucion_pedidos= productos_por_pedido.value_counts().sort_index()
distribucion_pedidos.plot(
    title="productos por pedidos",
    kind="bar",
    figsize=(30,5),
    xlabel="# de productos",
    ylabel="cantidad de pedidos",
    rot=360
)
plt.show()
print("Estadísticas de productos por pedido:")
print(f"En Promedio la cantidad de productos pedidos son: {productos_por_pedido.mean().astype(int)}")
print(f"La cantidad de pedidos pedidos con mas frecuencia son: {productos_por_pedido.mode()[0]}")
```
